```python
import joblib
import pandas as pd
import numpy as np
import os

# ---------------- CONFIG ----------------
MODEL_FILE = '/content/drive/MyDrive/Ml Project/antenna_s_model.joblib'
SAVE_FILE = "global_resonances.xlsx"

# Design bounds (mm)
MIN_N = 3.0
MAX_N = 22.0
MIN_WS = 10.0
MAX_WS = 24.0

# Grid resolution
N_STEPS = 500
WS_STEPS = 500

# Global frequency sweep (for resonance detection)
GLOBAL_FREQ_START = 2.0
GLOBAL_FREQ_END = 6.8
GLOBAL_FREQ_POINTS = 150

# Frequency resolution inside user band
BAND_FREQ_POINTS = 100
# ----------------------------------------


def load_model():
    if not os.path.exists(MODEL_FILE):
        print("Model file not found.")
        return None
    return joblib.load(MODEL_FILE)


def create_design_grid():
    n_grid = np.linspace(MIN_N, MAX_N, N_STEPS)
    ws_grid = np.linspace(MIN_WS, MAX_WS, WS_STEPS)
    return np.array(np.meshgrid(n_grid, ws_grid)).T.reshape(-1, 2)


def compute_and_save_global_resonance(model, design_pairs):
    print("Computing global resonances (2â€“6.8 GHz)...")

    freq_sweep = np.linspace(GLOBAL_FREQ_START,
                             GLOBAL_FREQ_END,
                             GLOBAL_FREQ_POINTS)

    total_designs = len(design_pairs)

    n_col = np.repeat(design_pairs[:, 0], GLOBAL_FREQ_POINTS)
    ws_col = np.repeat(design_pairs[:, 1], GLOBAL_FREQ_POINTS)
    freq_col = np.tile(freq_sweep, total_designs)

    predict_df = pd.DataFrame({
        'n': n_col,
        'ws': ws_col,
        'frequency': freq_col
    })

    predictions = model.predict(predict_df)
    s_matrix = predictions.reshape(total_designs, GLOBAL_FREQ_POINTS)

    best_s = np.min(s_matrix, axis=1)
    best_freq_idx = np.argmin(s_matrix, axis=1)
    best_freq = freq_sweep[best_freq_idx]

    result_df = pd.DataFrame({
        'n': design_pairs[:, 0],
        'ws': design_pairs[:, 1],
        'best_s': best_s,
        'best_freq': best_freq
    })

    result_df.to_excel(SAVE_FILE, index=False)
    print(f"Global resonances saved to {SAVE_FILE}")

    return result_df


def load_or_compute_resonance_table(model):
    if os.path.exists(SAVE_FILE):
        print("Loading saved global resonance table...")
        return pd.read_excel(SAVE_FILE)
    else:
        design_pairs = create_design_grid()
        return compute_and_save_global_resonance(model, design_pairs)


def compute_band_minimum(model, design_pairs, f_start, f_end):

    freq_band = np.linspace(f_start, f_end, BAND_FREQ_POINTS)
    total_designs = len(design_pairs)

    n_col = np.repeat(design_pairs[:, 0], BAND_FREQ_POINTS)
    ws_col = np.repeat(design_pairs[:, 1], BAND_FREQ_POINTS)
    freq_col = np.tile(freq_band, total_designs)

    predict_df = pd.DataFrame({
        'n': n_col,
        'ws': ws_col,
        'frequency': freq_col
    })

    predictions = model.predict(predict_df)
    s_matrix = predictions.reshape(total_designs, BAND_FREQ_POINTS)

    band_min_s = np.min(s_matrix, axis=1)

    return band_min_s


def main():

    model = load_model()
    if model is None:
        return

    # Load or compute global resonance table
    resonance_df = load_or_compute_resonance_table(model)

    design_pairs = resonance_df[['n', 'ws']].values
    global_best_s = resonance_df['best_s'].values
    global_best_freq = resonance_df['best_freq'].values

    while True:
        try:
            print("\nEnter desired frequency band:")
            f_start = float(input("Start (GHz): "))
            f_end = float(input("End (GHz): "))

            if f_start >= f_end:
                print("Start must be less than End.")
                continue

            if f_start < GLOBAL_FREQ_START or f_end > GLOBAL_FREQ_END:
                print(f"Band must lie within {GLOBAL_FREQ_START}â€“{GLOBAL_FREQ_END} GHz")
                continue

            # Step 1: Check if resonance lies inside band
            resonance_mask = (global_best_freq >= f_start) & (global_best_freq <= f_end)

            if np.any(resonance_mask):

                valid_indices = np.where(resonance_mask)[0]
                best_index = valid_indices[np.argmin(global_best_s[resonance_mask])]

                print("\n----- BEST RESONANT DESIGN -----")
                print(f"n  = {design_pairs[best_index][0]:.4f} mm")
                print(f"ws = {design_pairs[best_index][1]:.4f} mm")
                print(f"S  = {global_best_s[best_index]:.4f} dB")
                print(f"Resonant Frequency = {global_best_freq[best_index]:.4f} GHz")
                print("--------------------------------")

            else:
                print("\nNo resonance inside this band.")
                print("Computing lowest S only inside band...")

                band_min_s = compute_band_minimum(
                    model,
                    design_pairs,
                    f_start,
                    f_end
                )

                best_index = np.argmin(band_min_s)

                print("\n----- BEST DESIGN IN BAND -----")
                print(f"n  = {design_pairs[best_index][0]:.4f} mm")
                print(f"ws = {design_pairs[best_index][1]:.4f} mm")
                print(f"Minimum S in band = {band_min_s[best_index]:.4f} dB")
                print("--------------------------------")

        except ValueError:
            print("Enter numeric values only.")
        except KeyboardInterrupt:
            print("\nExiting.")
            break


if __name__ == "__main__":
    main()
```
