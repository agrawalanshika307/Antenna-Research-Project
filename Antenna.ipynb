```python
import joblib
import pandas as pd
import os

MODEL_FILE = 'antenna_s_model.joblib'

def load_model():
    if not os.path.exists(MODEL_FILE):
        return None
    try:
        return joblib.load(MODEL_FILE)
    except:
        return None

def main():
    model = load_model()
    if model is None:
        return

    while True:
        try:
            f_input = input("Frequency (GHz): ")
            if f_input.lower() == 'q':
                break

            n_input = input("Slot Length n (mm): ")
            if n_input.lower() == 'q':
                break

            ws_input = input("Slot Width ws (mm): ")
            if ws_input.lower() == 'q':
                break

            f_val = float(f_input)
            n_val = float(n_input)
            ws_val = float(ws_input)

            df = pd.DataFrame(
                [[n_val, ws_val, f_val]],
                columns=['n', 'ws', 'frequency']
            )

            pred = model.predict(df)[0]

            print(f"S-parameter: {pred:.4f} dB")
            print("Good match" if pred < -10 else "Poor match")

        except ValueError:
            pass
        except KeyboardInterrupt:
            break

if __name__ == "__main__":
    main()
```

```python
import joblib
import pandas as pd
import numpy as np
import os
import time

MODEL_FILE = 'antenna_s_model.joblib'
DATASET_FILE = 'data.csv'

N_STEPS = 500
WS_STEPS = 500
FREQ_TEST_POINTS = 10

def load_resources():
    if not os.path.exists(MODEL_FILE) or not os.path.exists(DATASET_FILE):
        return None, None
    try:
        return joblib.load(MODEL_FILE), pd.read_csv(DATASET_FILE)
    except:
        return None, None

def run_optimization(model, df):
    try:
        freq_start = float(input("Start frequency (GHz): "))
        freq_end = float(input("End frequency (GHz): "))
        if freq_start >= freq_end:
            return

        min_n, max_n = df['n'].min(), df['n'].max()
        min_ws, max_ws = df['ws'].min(), df['ws'].max()

        n_grid = np.linspace(min_n, max_n, N_STEPS)
        ws_grid = np.linspace(min_ws, max_ws, WS_STEPS)

        design_pairs = np.array(np.meshgrid(n_grid, ws_grid)).T.reshape(-1, 2)
        total_designs = len(design_pairs)

        test_frequencies = np.linspace(freq_start, freq_end, FREQ_TEST_POINTS)

        start_time = time.time()

        n_col = np.repeat(design_pairs[:, 0], FREQ_TEST_POINTS)
        ws_col = np.repeat(design_pairs[:, 1], FREQ_TEST_POINTS)
        freq_col = np.tile(test_frequencies, total_designs)

        predict_df = pd.DataFrame({
            'n': n_col,
            'ws': ws_col,
            'frequency': freq_col
        })

        predictions = model.predict(predict_df)
        s_param_matrix = predictions.reshape(total_designs, FREQ_TEST_POINTS)
        worst_s = np.max(s_param_matrix, axis=1)

        mask = worst_s < -10
        if not np.any(mask):
            return

        designs = design_pairs[mask]
        margins = worst_s[mask]

        results = sorted(zip(designs, margins), key=lambda x: x[1])

        end_time = time.time()

        for i, (design, margin) in enumerate(results[:5], 1):
            print(f"{i} {design[0]:.4f} {design[1]:.4f} {margin:.4f}")

        best = results[0][0]
        print(f"BEST {best[0]:.4f} {best[1]:.4f}")
        print(f"TIME {end_time - start_time:.2f}s")

    except:
        pass

if __name__ == "__main__":
    model, data = load_resources()
    if model is not None and data is not None:
        run_optimization(model, data)
```
